```python
from pyrite import _utils as ut
from pyrite.geoid import GeoidHeight as gh
import numpy as np
import scipy.io
from mpl_toolkits.basemap import Basemap

def ocean_mask(latlist,lonlist):
    row,col = latlist.shape
    mask = np.zeros((row,col))
    for i in range(row):
        for j in range(col):
            lat0 = latlist[i,j]
            lon0 = lonlist[i,j]
            if bm.is_land(lon0, lat0):
                mask[i,j] = 1
            
    return mask

def get_geoid(lat,lon):
    
    row,col = lat.shape
    lat = np.asarray(lat)
    lon = np.asarray(lon)
    lat0 = lat.flatten()
    lon0 = lon.flatten()  
    obj = gh()
    geoidH = []
    for i in range(len(lat0)):
        N0 = gh.get(obj, lat0[i], lon0[i], cubic=True)
        geoidH.append(N0)    
    geoidH = np.asarray(geoidH)
    geoidH = geoidH.reshape(row,col)
    return geoidH

bm = Basemap()   # default: projection='cyl'
print(bm.is_land(99.675, 13.104))  #True
print(bm.is_land(100.539, 13.104))  #False

cdic = ut.initconst()
fname0 = '/Users/caoy0a/Desktop/ERA-5_N60_N70_W25_W10_20160604_19.grb'

lvls,latlist,lonlist,gph,tmp,vpr = ut.get_ecmwf('ERA5',fname0,cdic, humidity='Q')
print(gph.shape)
print(gph[:,1,1])

hgt = np.linspace(cdic['minAltP'], gph.max().round(), cdic['nhgt'])
Presi,Tempi,Vpri = ut.intP2H(lvls,hgt,gph,tmp,vpr,cdic,verbose=False)

#scipy.io.savemat('/Users/caoy0a/Desktop/pyrite_test/latlist.mat', mdict={'latlist': latlist})
#scipy.io.savemat('/Users/caoy0a/Desktop/pyrite_test/lonlist.mat', mdict={'lonlist': lonlist})
#scipy.io.savemat('/Users/caoy0a/Desktop/pyrite_test/Pvapor.mat', mdict={'Pvapor': Vpri})
#scipy.io.savemat('/Users/caoy0a/Desktop/pyrite_test/Pressure.mat', mdict={'Pressure': Presi})
#scipy.io.savemat('/Users/caoy0a/Desktop/pyrite_test/Template.mat', mdict={'Template': Tempi})
#scipy.io.savemat('/Users/caoy0a/Desktop/pyrite_test/hgt.mat', mdict={'hgt': hgt})

print(latlist.shape)
print(latlist)
print(lonlist)

lonlist = lonlist - 360
mask = ocean_mask(latlist,lonlist)
#scipy.io.savemat('/Users/caoy0a/Desktop/pyrite_test/mask.mat', mdict={'mask': mask})

lat0 = latlist.flatten()
lon0 = lonlist.flatten()

print(bm.is_land(-18,65))

obj = gh()
h1 = gh.get(obj, lat0[0], lon0[0], cubic=True)
h0 = gh.get(obj, lat0[1], lon0[1], cubic=True)


row,col = latlist.shape
hh = get_geoid(latlist,lonlist)
print(hh)

#print(latlist.shape)
```
